```python
'''Дан файл HR.csv с данными по опросу уровня удовлетворенности сотрудниками работой.
Признаки:
    1.satisfaction_level  - Уровень удовлетворенности работой
    2.Last_evaluation  - Время с момента последней оценки в годах
    3.number_projects  - Количество проектов, выполненных за время работы
    4.average_monthly_hours  - Среднее количество часов на рабочем месте в месяц
    5.time_spend_company  - Стаж работы в компании в годах
    6.work_accident  - Происходили ли несчастные случаи на рабочем месте с сотрудником
    7.left  - уволился ли сотрудник
    8.promotion_last_5years  - повышался ли сотрудник за последние пять лет
    9.department  - отдел в котором работает сотрудник
    10.salary  - относительный уровень зарплаты'''
```

```python
import numpy as np
np.set_printoptions(precision=4)
import pandas as pd
from pandas import ExcelWriter

data_HR = pd.read_csv('HR.csv',#загружаем данные из файла
                       sep=',', encoding='utf-8'
                       )
data_HR.head()
```

```python
data_ = data_HR.copy()
data_.drop(['left','department','salary'], axis=1, inplace=True)#удаляем ненужные стлобцы
data = data_.to_numpy()#преобразуем в массив numpy
data.shape
```

```python
data_class_arr = data_HR.left
data_class = data_class_arr.to_numpy()
data_class
```

```python
# разделяем выборку на тренировочную и тестовую
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(data, data_class, test_size=0.25)
```

```python
# импортируем LDA-функционал
from sklearn.discriminant_analysis import LinearDiscriminantAnalysis
```

```python
lda = LinearDiscriminantAnalysis()
```

```python
# обучаем данные
lda.fit(X_train, y_train)
```

```python
# делаем прогноз на тестовой выборке
lda.predict(X_test)
```

```python
# смотрим разницу факта и прогноза
result = pd.DataFrame([y_test, lda.predict(X_test)]).T
result
```

```python
# расчет точности - отношение верных прогнозов к общему количеству позиций
# является хорошей метрикой для сбалансированных классов 
# (наблюдений в категориях примерно одинаковое количество)

from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, lda.predict(X_test))
```
